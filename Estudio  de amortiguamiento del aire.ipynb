```python
#Función para leer los archivos Txt y se deshace de las líneas encabezado
def leer_txt(ruta):
    DatosT = open(ruta)
    DatosT = DatosT.readlines()
    DatosT.pop(0)
    DatosT.pop(0)
    DatosT.pop(0)
    DatosT.pop(0)
    DatosT.pop(0)
    return DatosT
```

```python
#Funciones para separar la información y almacenarla en un par de Array
def formato(x):
    x1 = np.array(x)
    x2 = pd.Series(x1)
    x3 = x2.apply(separar)
    return x3
```

```python
def separar(x):
    return x.split()
```

```python
def anum(x):
    res = [float(ele) for ele in x] 
    return res
```

```python
#Función para almacenar toda la información en un Dataframe
def Cons_DF(Dpn,p):
    DatosTr = pd.DataFrame(columns='t TSR0.2T TSR0.3T TSR0.4T TSR0.5T TSR0.6T TSR0.7T TSR0.8T TSR0.2Tn TSR0.3Tn TSR0.4Tn TSR0.5Tn TSR0.6Tn TSR0.7Tn TSR0.8Tn TSR0.2Tp TSR0.3Tp TSR0.4Tp TSR0.5Tp TSR0.6Tp TSR0.7Tp TSR0.8Tp '.split())
    Cont = 0
    for x in Dpn:
        try:
            DatosTr.at[Cont,'TSR0.2T'] = x[1+p] + x[2+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.2T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.3T'] = x[3+p] + x[4+p] 
        except IndexError:
            DatosTr.at[Cont,'TSR0.3T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.4T'] = x[5+p] + x[6+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.4T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.5T'] = x[7+p] + x[8+p] 
        except IndexError:
            DatosTr.at[Cont,'TSR0.5T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.6T'] = x[9+p] + x[10+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.6T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.7T'] = x[11+p] + x[12+p] 
        except IndexError:
            DatosTr.at[Cont,'TSR0.7T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.8T'] = x[13+p] + x[14+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.8T'] = 0
        try:
            DatosTr.at[Cont,'TSR0.2Tn'] = Dpn[Cont][1+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.2Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.3Tn'] = Dpn[Cont][3+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.3Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.4Tn'] = Dpn[Cont][5+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.4Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.5Tn'] = Dpn[Cont][7+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.5Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.6Tn'] = Dpn[Cont][9+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.6Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.7Tn'] = Dpn[Cont][11+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.7Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.8Tn'] = Dpn[Cont][13+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.8Tn'] = 0
        try:
            DatosTr.at[Cont,'TSR0.2Tp'] = Dpn[Cont][2+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.2Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.3Tp'] = Dpn[Cont][4+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.3Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.4Tp'] = Dpn[Cont][6+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.4Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.5Tp'] = Dpn[Cont][8+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.5Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.6Tp'] = Dpn[Cont][10+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.6Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.7Tp'] = Dpn[Cont][12+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.7Tp'] = 0
        try:
            DatosTr.at[Cont,'TSR0.8Tp'] = Dpn[Cont][14+p]
        except IndexError:
            DatosTr.at[Cont,'TSR0.8Tp'] = 0
        Cont = Cont + 1
    return DatosTr
```

```python
#Función de eddy para calcular el CP y la velocidad de giro en estado estacionario
def Calc_CP(DatosTr,inicio,paso,final,V,R_r,n_m,n_c):
    TSR_sweep = np.arange(inicio,final+paso,paso)
    wind_velocity = V
    radio_rotor = 0.25
    num_muestras = 150
    num_ciclos = 3
    num_tsr = 7
    pi = 3.1416
    #Evaluar la expresión de Torque y calcular vectores de tiempo
    Tiempo = np.zeros((num_tsr,150))
    for i in np.arange(0,7,1):
        Tiempo[i] = np.linspace(0,(1/(((TSR_sweep[i]*wind_velocity)/radio_rotor)/(2*pi)))*num_ciclos,num_muestras)
    
    TM = (1/(((0.2*wind_velocity)/radio_rotor)/(2*pi)))*num_ciclos
    
    #Integrar y calcular el Torque promedio
    avgTorque = np.zeros((num_tsr,1))
    angVel = np.zeros((num_tsr,1))
    Cont = 0
    for i in 'TSR0.2T TSR0.3T  TSR0.4T TSR0.5T  TSR0.6T TSR0.7T  TSR0.8T'.split():#Esta línea está modificada debido a que en este experimento unicamente convergieron con TSR 0.2 y 0.3
        Te = (1/(((TSR_sweep[Cont]*wind_velocity)/radio_rotor)/(2*pi)))*num_ciclos
        NM = 0
        val = 0
        for p in DatosTr[i]:
            if p == val:
                break
            else:
                val = p
                NM = NM+1
        avgTorque[Cont] = np.sum(DatosTr[i].values[1:NM-1])/((NM-1)-1)
        angVel[Cont] = ((TSR_sweep[Cont]*wind_velocity)/radio_rotor) #Originalmente en rad/s. 
        #Para convertir a rev/s dividir por --> /(2*pi)
        Cont = Cont+1
    
    
    A = 0.5*0.5
    rho = 1.225 #1.225 --> paper de generación de calor con Sav y paper de Ortega. --- 0.3 Anterior.
    P_wind = 0.5*rho*A*float(wind_velocity)**float(3)
    P_rot = np.zeros((num_tsr,1))
    for i in np.arange(0,7,1):
        P_rot[i] =   avgTorque[i] * angVel[i]
    
    #Calcular CP para cada experimento del DOE y almacenarlo
    CP = P_rot/P_wind 
    return CP,angVel,P_wind,Tiempo
```
